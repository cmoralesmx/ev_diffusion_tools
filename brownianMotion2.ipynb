# Brownian motion

Simulation using one particle and two dimmensions  

Firstly, we setup our particle.

```python
from scipy import stats
from scipy.stats import norm
from itertools import cycle
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
steps = 1000
particle = {'x': np.append(0.0, norm.rvs(size=steps)), 'y': np.append(0.0, norm.rvs(size=steps))}
```

```python
plt.plot(np.cumsum(particle['x']), np.cumsum(particle['y']))
```

# Multiple EVs

Extending our model for 1 < N < 1000

```python
N = 30
```

```python
particles = []
for idx in range(N):
    particles.append({'x':np.append(0.0, norm.rvs(size=steps)), 'y':np.append(0.0, norm.rvs(size=steps))}) 
#particles
```

```python
for idx in range(N):
    plt.plot(np.cumsum(particles[idx]['x']), np.cumsum(particles[idx]['y']))
```
